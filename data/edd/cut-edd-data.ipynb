```python
def filter_edd(edd, regions_dict):
    # separate dataframes
    ba = edd.loc[edd['Regions'] == 'Bay Area']
    cc = edd.loc[edd['Regions'] == 'Central Coast']
    cv = edd.loc[edd['Regions'] == 'Central Valley']
    ie = edd.loc[edd['Regions'] == 'Inland Empire']
    la = edd.loc[edd['Regions'] == 'Los Angeles']
    oc = edd.loc[edd['Regions'] == 'Orange']
    rc = edd.loc[edd['Regions'] == 'Redwood Coast']
    sac = edd.loc[edd['Regions'] == 'Sacramento']
    sd = edd.loc[edd['Regions'] == 'San Diego-Imperial']
    shas = edd.loc[edd['Regions'] == 'Shasta / Cascades']
    sn = edd.loc[edd['Regions'] == 'Sierra Nevada']
    
    # filter out series codes
    ba = ba.loc[ba['Series Code'].isin(regions_dict['Bay Area'])]
    cc = cc.loc[cc['Series Code'].isin(regions_dict['Central Coast'])]
    cv = cv.loc[cv['Series Code'].isin(regions_dict['Central Valley'])]
    ie = ie.loc[ie['Series Code'].isin(regions_dict['Inland Empire'])]
    la = la.loc[la['Series Code'].isin(regions_dict['Los Angeles'])]
    oc = oc.loc[oc['Series Code'].isin(regions_dict['Orange'])]
    rc = rc.loc[rc['Series Code'].isin(regions_dict['Redwood Coast'])]
    sac = sac.loc[sac['Series Code'].isin(regions_dict['Sacramento'])]
    sd = sd.loc[sd['Series Code'].isin(regions_dict['San Diego-Imperial'])]
    shas = shas.loc[shas['Series Code'].isin(regions_dict['Shasta / Cascades'])]
    sn = sn.loc[sn['Series Code'].isin(regions_dict['Sierra Nevada'])]
    
    dfs = [ba, cc, cv, ie, la, oc, rc, sac, sd, shas, sn]
    final_edd = pd.concat(dfs)
    
    return final_edd
```
